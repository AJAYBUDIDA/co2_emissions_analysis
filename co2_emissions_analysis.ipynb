```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style='whitegrid')

# 2. Load the Dataset
df = pd.read_csv('global.csv')

# 3. Preview the Data
print("First 5 rows:")
print(df.head())

print("\nShape of dataset:", df.shape)
print("\nColumns:", df.columns.tolist())

# 4. Check data types and null values
print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

# 5. Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())

# 6. Total CO2 emissions by year
yearly_emissions = df.groupby('Year')['Total'].sum().reset_index()

# Plot total emissions over years
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_emissions, x='Year', y='Total', marker='o')
plt.title('Global CO₂ Emissions Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Emissions (Million Metric Tons)')
plt.tight_layout()
plt.show()

# 7. Top 10 Countries with Highest Total Emissions
country_emissions = df.groupby('Country')['Total'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=country_emissions.values, y=country_emissions.index, palette='Reds_r')
plt.title('Top 10 Countries with Highest CO₂ Emissions (Total)')
plt.xlabel('Total Emissions (Million Metric Tons)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

# 8. Pie Chart: Contribution of Top 5 Countries (in 2014 or latest year)
latest_year = df['Year'].max()
latest_data = df[df['Year'] == latest_year].sort_values(by='Total', ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.pie(latest_data['Total'], labels=latest_data['Country'], autopct='%1.1f%%', startangle=140)
plt.title(f'Top 5 CO₂ Emitting Countries in {latest_year}')
plt.axis('equal')
plt.tight_layout()
plt.show()

# 9. Heatmap: Correlation between emission types
emission_cols = ['Gas Fuel', 'Liquid Fuel', 'Solid Fuel', 'Cement', 'Gas Flaring', 'Total']
plt.figure(figsize=(10, 6))
sns.heatmap(df[emission_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Different Emission Types')
plt.tight_layout()
plt.show()
```
